```python
import requests
import os
from urllib.request import urlopen
import xmltodict
import csv
import pandas as pd
import itertools
```

```python
def CreateCSV(path, classname):

  svs_list = [svs_file for svs_file in os.listdir(path) if svs_file.endswith(".tiff")]

  # read images
  svs_files = []
  for svs in svs_list:
    svs_files.append(svs[0:12])
  svs_files = list(dict.fromkeys(svs_files))
 
  # api-endpoint
  URL = 'https://api.gdc.cancer.gov/files?filters={"op":"and","content":[{"op":"=","content":{"field":"files.data_format","value":["BCR XML"]}},{"op":"=","content":{"field":"cases.project.project_id","value":["TCGA-' + classname + '"]}},{"op":"=","content":{"field":"data_type","value":["Clinical Supplement"]}}]}&fields=file_id,associated_entities.entity_submitter_id&pretty=TRUE&from=0&size=99999'
  URL = URL.replace(" ", "%20")
  r = requests.get(url = URL)
  myfile = r.json()

  # patient - file
  tiffdic = {}
  for i in range(len(myfile['data']['hits'])):
    file_id = myfile['data']['hits'][i]
    patient_id = file_id['associated_entities'][0]['entity_submitter_id']
    if patient_id in svs_files:
      tiffdic[patient_id] = file_id['id']

  # for testing
  # tiffdic = dict(itertools.islice(tiffdic.items(), 10))

  # patient - age
  mydict = {}
  for patient_id, file_id in tiffdic.items():
    try:
      URL = "https://api.gdc.cancer.gov/data/{}".format(file_id)
      data = urlopen(URL).read()
      data = xmltodict.parse(data)
      age = data['skcm:tcga_bcr']['skcm:patient']['clin_shared:age_at_initial_pathologic_diagnosis']['#text']
      mydict[patient_id] = age  
    except:
      continue

  csvname = classname + '.csv'
  with open(csvname, 'w') as csv_file:  
    writer = csv.writer(csv_file)
    writer.writerow(["patient id", "age"])
    for key, value in mydict.items():
       writer.writerow([key, value])
```

```python
path = '/content/gdrive/MyDrive/TCGA/TCGA-SKCM'
CreateCSV(path, "SKCM")
```
