```python
import glob

def flatten(s_path) : 

  svs_list = [svs_dir for svs_dir in os.listdir(s_path)]

  suff_counter = 0
  
  for s in svs_list:

    try:

      path = s_path+'/'+s+'//**//*.png'

      tile_dir = "/content/gdrive/MyDrive/project/TCGA-LIHC/tiles/"
      
      for filepath in glob.glob(path, recursive=True):

        if filepath.endswith(".png") :

          i = 0

          filepath_arr = filepath.split("/")

          name = filepath_arr[-1]

          if "tilecrossed" in name : 
            continue

          name = name[:-4] +"_"+ str(i)

          image = Image.open(filepath)
          print(image)
          suff_counter = suff_counter + 1
          i = i + 1

          image.save(tile_dir+name+".tiff")

          print(str(suff_counter) + 'th Tiff generated with name: ', tile_dir+name+".tiff")

    except:
      print("Exception occurred")
      continue
  return suff_counter

s_path = '/content/gdrive/MyDrive/project/TCGA-LIHC/outputs'
j = flatten(s_path)

print(j)
```
