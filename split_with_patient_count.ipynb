```python
import os
from collections import Counter
import glob
import shutil
```

## Get class name

```python
classname = "COAD"
```

```python
path = '/content/gdrive/MyDrive/TCGA/TCGA-' + classname + '/'
```

```python
t_list = [svs_file for svs_file in os.listdir(path) if svs_file.endswith(".tiff")]
print(t_list)
```

```python
myfiles = []
for svs in t_list:
  myfiles.append(svs[0:12])
```

```python
tiff_count = len(myfiles) 
print(tiff_count)
```

```python
# Counter(myfiles)
```

## Patient number

```python
singlefiles = list(dict.fromkeys(myfiles))
print(len(singlefiles))
```

## Get n patients

```python
train_num = 150
validation_num = 18
test_num = 18
```

```python
train = singlefiles[0:train_num]
validation = singlefiles[train_num: train_num + validation_num]
test = singlefiles[train_num + validation_num:]

print(len(train))
print(len(validation))
print(len(test))
```

```python
traindata = [t_file for t_file in t_list if t_file[0:12] in train]
validationdata = [t_file for t_file in t_list if t_file[0:12] in validation]
testdata = [t_file for t_file in t_list if t_file[0:12] in test]

print(len(traindata))
print(len(validationdata))
print(len(testdata))

for row in t_list:
  if row in validationdata and row in testdata:
    print(row)
```

```python
print(traindata)
```

## Get patient images

```python
src_dir = '/content/gdrive/MyDrive/TCGA/TCGA-' + classname + '/'

t_count = 0
v_count = 0
test_count = 0

for tiff_file in glob.iglob(os.path.join(src_dir, "*.tiff")):

  t_file = tiff_file[39:]
  
  if t_file in traindata:
    dst_dir = '/content/gdrive/MyDrive/new-data1/train/TCGA-' + classname + '/'
    shutil.copy(tiff_file, dst_dir)
    t_count += 1

  elif t_file in validationdata:
    dst_dir = '/content/gdrive/MyDrive/new-data1/validation/TCGA-' + classname + '/'
    shutil.copy(tiff_file, dst_dir)
    v_count += 1
  
  elif t_file in testdata:
    dst_dir = '/content/gdrive/MyDrive/new-data1/test/TCGA-' + classname + '/'
    shutil.copy(tiff_file, dst_dir)
    test_count += 1

  else:
    print(tiff_file)

print(t_count, v_count, test_count)
```
